IMPORTING THE LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from joblib import dump
```

LOADING THE DATASET

```python
df = pd.read_csv('players_15.csv')
```

```python
df.head()
```

```python
df.info()
```

CHECKING FOR PASSING-RELATED COLUMNS

```python
def find_col(df, names):
    """Find first matching existing column for candidates (case-insensitive, no punctuation)."""
    norm = { ''.join(ch for ch in c.lower() if ch.isalnum()): c for c in df.columns }
    for n in names:
        key = ''.join(ch for ch in n.lower() if ch.isalnum())
        if key in norm:
            return norm[key]
    return None

short_col = find_col(df, ["short_passing", "short pass", "shortpass"])
long_col  = find_col(df, ["long_passing", "long pass", "longpass"])
passing_col = find_col(df, ["passing", "pass"])
overall_col = find_col(df, ["overall", "ovr"])

print("Detected columns -> short:", short_col, " long:", long_col, " passing:", passing_col, " overall:", overall_col)
```

CREATING AVERAGE_PASS_SCORE

```python
if short_col and long_col:
    df["avg_pass_score"] = df[[short_col, long_col]].mean(axis=1)
    print("avg_pass_score = mean(short, long)")
elif passing_col:
    df["avg_pass_score"] = df[passing_col]
    print("avg_pass_score = passing")
elif overall_col:
    df["avg_pass_score"] = df[overall_col]
    print("avg_pass_score = overall (proxy)")
else:
    raise ValueError("No passing/overall columns found. Check your CSV.")
```

DEFINING THE THRESHOLD

```python
THRESHOLD = 70
df["pass_completed"] = (df["avg_pass_score"] > THRESHOLD).astype(int)
```

VISUALIZATION

```python
print("Target counts:\n", df["pass_completed"].value_counts())
plt.figure(figsize=(5,3))
sns.countplot(x="pass_completed", data=df)
plt.title("Target distribution (0 = not completed, 1 = completed)")
plt.show()
```

SELECTING THE FEATURES WE NEED

```python
candidate_features = [
    "vision", "ball_control", "dribbling", "positioning", "composure",
    "reactions", "stamina", "short_passing", "long_passing", "passing",
    "agility", "balance", "overall", "potential", "age", "height_cm", "weight_kg"
]

selected = []
for name in candidate_features:
    c = find_col(df, [name])
    if c and c not in selected:
        selected.append(c)

# Fallback: pick some numeric columns if too few detected
if len(selected) < 4:
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    numeric_cols = [c for c in numeric_cols if c not in ("pass_completed", "avg_pass_score")]
    selected = numeric_cols[:8]
    print("Fallback numeric selection used.")
print("Selected features:", selected)
```

```python
df.info()
```

FIXING MISSING VALUES

```python
X = df[selected].copy()
y = df["pass_completed"].copy()

# Fill numeric missing values with median
for col in X.columns:
    if X[col].isnull().any():
        X[col] = X[col].fillna(X[col].median())
```

SPLITTING THE DATASET

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
```

TRAINING THE MODEL

```python
clf = DecisionTreeClassifier(max_depth=4, random_state=42)
clf.fit(X_train, y_train)
```

EVALUATING THE MODEL BASED ON ACCURACY

```python
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"Test accuracy: {acc:.4f}")
```

BASIC VISUALIZATION

```python
plt.figure(figsize=(16,7))
plot_tree(clf, feature_names=X.columns, class_names=["NotCompleted","Completed"],
          filled=True, rounded=True, fontsize=10)
plt.title("Decision Tree (max_depth=4)")
plt.show()
```

PREDICTING WITH NEW FEATURES

```python
players = [
    [80, 28, 182, 76, 78, 82, 79],  # player 1
    [65, 21, 175, 68, 60, 62, 61]   # player 2
]

new_data = pd.DataFrame(players, columns=X.columns)
predictions = clf.predict(new_data)
print(predictions)
```

## TASK 1: Feature Engineering and Data Improvement

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

print("Dataset preview:")
display(df.head())

# ------------------------------------------------
# HANDLE MISSING DATA
# ------------------------------------------------
print("\nMissing values before cleaning:")
print(df.isnull().sum())

# Fill numeric columns with mean, categorical with mode
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        df[col].fillna(df[col].mean(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)

print("\nMissing values handled successfully!")




```

```python
# ------------------------------------------------
# ADDing NEW FEATURES
# ------------------------------------------------
# Adjust column names to match the dataset
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

if len(num_cols) >= 2:
    # new feature 1: sum of two numeric columns
    df['feature_sum'] = df[num_cols[0]] + df[num_cols[1]]

    # \new feature 2: ratio of same two columns (avoid division by zero)
    df['feature_ratio'] = df[num_cols[0]] / (df[num_cols[1]] + 1e-5)

if len(num_cols) >= 3:
    # new feature 3: mean of three numeric columns
    df['feature_mean3'] = df[num_cols[:3]].mean(axis=1)

print("\nNew features added successfully!")
print(df.head())
```

```python
# ------------------------------------------------
# FEATURE SELECTION
# ------------------------------------------------
# --- Option A: Correlation method
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()

# Drop highly correlated features (threshold > 0.9)
threshold = 0.9
corr_abs = corr_matrix.abs()
upper = corr_abs.where(np.triu(np.ones(corr_abs.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > threshold)]
df_reduced = df.drop(columns=to_drop, errors='ignore')

print("\nDropped highly correlated features:", to_drop)
print("Remaining features:", df_reduced.shape[1])
```

```python
# ------------------------------------------------
# SHORT COMPARISON SUMMARY
# ------------------------------------------------
print("\nTASK 1 COMPLETE!")
print(f"Original features: {len(num_cols)}")
print(f"New features added: 2–3")
print(f"Features removed due to correlation: {len(to_drop)}")
print(f"Final dataset shape: {df_reduced.shape}")

y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"Test accuracy: {acc:.4f}")
```

The model before feature engineering had an accuracy of 100%. After feature engineering, it still maintained an accuracy of 100% this is because we had done feature engineering initially.

## Task 3: Model Explainability

Feature Importance (Simple Explainability)

```python
# Task 3: Model Explainability - Logistic Regression Solution
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

print("=== MODEL EXPLAINABILITY ANALYSIS USING LOGISTIC REGRESSION ===\n")

# Data preprocessing
print("Step 1: Data Preprocessing...")
X_clean = X_train.dropna(axis=1, how='all')
X_clean = X_clean.dropna()
y_clean = y_train[X_clean.index]
print(f"Final dataset: {X_clean.shape} features, {len(y_clean)} samples")

# Standardize features for Logistic Regression (important for coefficient interpretation)
print("\nStep 2: Standardizing features...")
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_clean)
X_scaled = pd.DataFrame(X_scaled, columns=X_clean.columns, index=X_clean.index)

# Train Logistic Regression model
print("Step 3: Training Logistic Regression Model...")
model = LogisticRegression(random_state=42, max_iter=1000)
model.fit(X_scaled, y_clean)

# Get feature importance using absolute coefficients
print("Step 4: Calculating feature importance...")
features = X_clean.columns
coefficients = model.coef_[0]  # Get coefficients for the first class
importances = np.abs(coefficients)  # Use absolute values for importance

# Create feature importance dataframe
imp_df = pd.DataFrame({'Feature': features, 'Importance': importances, 'Coefficient': coefficients})
imp_df = imp_df.sort_values('Importance', ascending=False)

print("\n" + "="*60)
print("FEATURE IMPORTANCE ANALYSIS - LOGISTIC REGRESSION")
print("="*60)

# VISUALIZATION 1: Top 10 Feature Importance Bar Chart
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
top10 = imp_df.head(10)
colors = ['green' if coef > 0 else 'red' for coef in top10['Coefficient']]
plt.barh(top10['Feature'], top10['Importance'], color=colors, alpha=0.7)
plt.xlabel('Absolute Coefficient (Importance)')
plt.title('Top 10 Most Important Features\n(Green=Positive, Red=Negative Impact)')
plt.gca().invert_yaxis()

# VISUALIZATION 2: Feature Importance Distribution
plt.subplot(2, 2, 2)
plt.hist(imp_df['Importance'], bins=20, alpha=0.7, color='lightcoral')
plt.xlabel('Absolute Coefficient')
plt.ylabel('Frequency')
plt.title('Distribution of Feature Importances')
plt.grid(True, alpha=0.3)

# VISUALIZATION 3: Cumulative Importance
plt.subplot(2, 2, 3)
cumulative_importance = imp_df['Importance'].cumsum()
plt.plot(range(1, len(cumulative_importance) + 1), cumulative_importance, 
         marker='o', linewidth=2, markersize=4)
plt.xlabel('Number of Features')
plt.ylabel('Cumulative Importance')
plt.title('Cumulative Feature Importance')
plt.grid(True, alpha=0.3)

# VISUALIZATION 4: Coefficient Signs (Positive vs Negative)
plt.subplot(2, 2, 4)
positive_coefs = (imp_df['Coefficient'] > 0).sum()
negative_coefs = (imp_df['Coefficient'] < 0).sum()
neutral_coefs = len(imp_df) - positive_coefs - negative_coefs

labels = ['Positive', 'Negative', 'Neutral']
sizes = [positive_coefs, negative_coefs, neutral_coefs]
colors = ['lightgreen', 'lightcoral', 'lightgray']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Direction of Feature Impact')

plt.tight_layout()
plt.show()

# VISUALIZATION 5: Detailed Top Features with Coefficients
plt.figure(figsize=(14, 8))
top15 = imp_df.head(15)

# Create horizontal bars with color coding
colors = ['green' if coef > 0 else 'red' for coef in top15['Coefficient']]
bars = plt.barh(top15['Feature'], top15['Importance'], color=colors, alpha=0.7)

# Add value labels with coefficient signs
for bar, importance, coef in zip(bars, top15['Importance'], top15['Coefficient']):
    width = bar.get_width()
    sign = '+' if coef > 0 else '-'
    plt.text(width + 0.001, bar.get_y() + bar.get_height()/2, 
             f'{width:.4f} ({sign})', ha='left', va='center', fontsize=9)

plt.xlabel('Absolute Coefficient Value (Importance)', fontsize=12)
plt.title('Top 15 Feature Importances with Direction of Impact\n(Green=Positive, Red=Negative)', fontsize=14)
plt.tight_layout()
plt.show()

# VISUALIZATION 6: Actual Coefficients Plot (Not Absolute)
plt.figure(figsize=(12, 6))
top15_actual = imp_df.head(15).sort_values('Coefficient', ascending=True)
colors = ['green' if coef > 0 else 'red' for coef in top15_actual['Coefficient']]
plt.barh(top15_actual['Feature'], top15_actual['Coefficient'], color=colors, alpha=0.7)
plt.axvline(x=0, color='black', linestyle='-', alpha=0.3)
plt.xlabel('Coefficient Value')
plt.title('Actual Coefficient Values\n(Positive increases probability, Negative decreases)')
plt.tight_layout()
plt.show()

# TOP 3 FEATURES ANALYSIS
print("\n" + "="*60)
print("TOP 3 FEATURES INFLUENCING PREDICTIONS")
print("="*60)

top3_features = imp_df.head(3)

for i, (idx, row) in enumerate(top3_features.iterrows(), 1):
    direction = "positive" if row['Coefficient'] > 0 else "negative"
    effect = "increases" if row['Coefficient'] > 0 else "decreases"
    print(f"\n{i}. **{row['Feature']}**")
    print(f"   • Importance Score: {row['Importance']:.4f}")
    print(f"   • Coefficient: {row['Coefficient']:.4f} ({direction})")
    print(f"   • Impact: A one standard deviation increase {effect} the prediction probability")
    print(f"   • Rank: #{i} most influential feature")

print(f"\n**Summary**: The top 3 features collectively account for " 
      f"{top3_features['Importance'].sum():.1%} of the model's total feature importance.")

# ADDITIONAL INSIGHTS
print("\n" + "="*60)
print("ADDITIONAL MODEL INTERPRETABILITY INSIGHTS")
print("="*60)

# Feature importance statistics
print(f"\nFeature Importance Statistics:")
print(f"   • Total features analyzed: {len(imp_df)}")
print(f"   • Features with positive impact: {(imp_df['Coefficient'] > 0).sum()}")
print(f"   • Features with negative impact: {(imp_df['Coefficient'] < 0).sum()}")
print(f"   • Most positive feature: {imp_df.loc[imp_df['Coefficient'].idxmax()]['Feature']} "
      f"(coef: {imp_df['Coefficient'].max():.4f})")
print(f"   • Most negative feature: {imp_df.loc[imp_df['Coefficient'].idxmin()]['Feature']} "
      f"(coef: {imp_df['Coefficient'].min():.4f})")

# Number of features needed for different coverage levels
print(f"\nFeature Coverage Analysis:")
for coverage in [0.5, 0.8, 0.9]:
    n_features = np.argmax(cumulative_importance >= coverage) + 1
    print(f"   • Features needed for {coverage:.0%} coverage: {n_features}")

# Model performance metrics
print(f"\nModel Information:")
print(f"   • Model Type: Logistic Regression")
print(f"   • Training samples: {len(X_scaled)}")
print(f"   • Features used: {len(features)}")
print(f"   • Convergence: {'Yes' if model.n_iter_ < model.max_iter else 'No (max iterations reached)'}")

print("\n" + "="*60)
print("MODEL EXPLAINABILITY SUMMARY - LOGISTIC REGRESSION")
print("="*60)
print("""
Key Advantages of Logistic Regression for Explainability:
• Coefficients directly interpretable as log-odds ratios
• Clear direction of impact (positive/negative coefficients)
• Linear relationship assumptions provide transparent reasoning
• Feature importance based on magnitude of standardized coefficients
""")
```

## Task 4: Result Visualization and Interpretation

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score

print("=== TASK 4: RESULT VISUALIZATION AND INTERPRETATION ===\n")

# First, let's train multiple models for comparison
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

# Data preparation (using your cleaned data)
X_clean = X_train.dropna(axis=1, how='all')
X_clean = X_clean.dropna()
y_clean = y_train[X_clean.index]

# Prepare test data similarly
X_test_clean = X_test[X_clean.columns].dropna()
y_test_clean = y_test[X_test_clean.index]

print("Training multiple models for comparison...")

# Define models
models = {
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
}

# Train models and store results
results = {}
for name, model in models.items():
    model.fit(X_clean, y_clean)
    y_pred = model.predict(X_test_clean)
    accuracy = accuracy_score(y_test_clean, y_pred)
    precision = precision_score(y_test_clean, y_pred, average='weighted')
    recall = recall_score(y_test_clean, y_pred, average='weighted')
    f1 = f1_score(y_test_clean, y_pred, average='weighted')
    
    results[name] = {
        'model': model,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'predictions': y_pred
    }

print("Models trained successfully!")

# VISUALIZATION 1: Model Accuracy Comparison
print("\n" + "="*50)
print("VISUALIZATION 1: Model Accuracy Comparison")
print("="*50)

plt.figure(figsize=(12, 6))

# Subplot 1: Accuracy Comparison
plt.subplot(1, 2, 1)
model_names = list(results.keys())
accuracies = [results[name]['accuracy'] for name in model_names]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

bars = plt.bar(model_names, accuracies, color=colors, alpha=0.8)
plt.ylabel('Accuracy Score')
plt.title('Model Accuracy Comparison', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.ylim(0, 1.0)

# Add value labels on bars
for bar, accuracy in zip(bars, accuracies):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{accuracy:.3f}', ha='center', va='bottom', fontweight='bold')

# Subplot 2: Multiple Metrics Comparison
plt.subplot(1, 2, 2)
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
x_pos = np.arange(len(model_names))
width = 0.2

for i, metric in enumerate(['accuracy', 'precision', 'recall', 'f1']):
    values = [results[name][metric] for name in model_names]
    plt.bar(x_pos + i*width, values, width, label=metric.title(), alpha=0.8)

plt.xlabel('Models')
plt.ylabel('Score')
plt.title('Multiple Metrics Comparison', fontsize=14, fontweight='bold')
plt.xticks(x_pos + 1.5*width, model_names, rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.ylim(0, 1.0)

plt.tight_layout()
plt.show()
```

Explanation: This comparison shows how different machine learning algorithms perform on our dataset. The left chart displays accuracy scores for easy model-to-model comparison, while the right chart provides a comprehensive view across multiple evaluation metrics including precision, recall, and F1-score.

```python
# VISUALIZATION 2: Confusion Matrix for Best Model
print("\n" + "="*50)
print("VISUALIZATION 2: Confusion Matrix Analysis")
print("="*50)

# Find best model based on accuracy
best_model_name = max(results, key=lambda x: results[x]['accuracy'])
best_model = results[best_model_name]['model']
y_pred_best = results[best_model_name]['predictions']

print(f"Best performing model: {best_model_name}")

# Create confusion matrix
cm = confusion_matrix(y_test_clean, y_pred_best)

plt.figure(figsize=(10, 4))

# Subplot 1: Standard Confusion Matrix
plt.subplot(1, 2, 1)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title(f'Confusion Matrix - {best_model_name}', fontweight='bold')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

# Subplot 2: Normalized Confusion Matrix
plt.subplot(1, 2, 2)
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
sns.heatmap(cm_normalized, annot=True, fmt='.2%', cmap='Greens', cbar=False)
plt.title(f'Normalized Confusion Matrix - {best_model_name}', fontweight='bold')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

plt.tight_layout()
plt.show()
```

Explanation: The confusion matrices show how well our best model classifies instances. The left matrix displays raw counts of correct and incorrect predictions, while the right normalized matrix shows percentages, making it easier to identify patterns in misclassifications across different classes.

```python
# VISUALIZATION 3: Detailed Evaluation Metrics Table
print("\n" + "="*50)
print("VISUALIZATION 3: Comprehensive Evaluation Metrics")
print("="*50)

# Create detailed metrics table
metrics_data = []
for name in model_names:
    metrics_data.append({
        'Model': name,
        'Accuracy': f"{results[name]['accuracy']:.4f}",
        'Precision': f"{results[name]['precision']:.4f}",
        'Recall': f"{results[name]['recall']:.4f}",
        'F1-Score': f"{results[name]['f1']:.4f}"
    })

metrics_df = pd.DataFrame(metrics_data)
print("Detailed Performance Metrics:")
print(metrics_df.to_string(index=False))

# Visual table with highlighting
plt.figure(figsize=(12, 3))
plt.axis('off')
plt.table(cellText=metrics_df.values,
          colLabels=metrics_df.columns,
          cellLoc='center',
          loc='center',
          bbox=[0, 0, 1, 1])
plt.title('Comprehensive Model Evaluation Metrics', fontsize=14, fontweight='bold', pad=20)
plt.show()

# Print classification report for best model
print(f"\nDetailed Classification Report for {best_model_name}:")
print(classification_report(y_test_clean, y_pred_best))
```

Explanation: This comprehensive metrics table provides a detailed comparison of all models across key evaluation metrics. The classification report further breaks down performance for each class, showing precision, recall, and F1-score at a granular level to identify any class-specific performance issues.

```python
# VISUALIZATION 4: Feature Importance Analysis
print("\n" + "="*50)
print("VISUALIZATION 4: Feature Importance Analysis")
print("="*50)

# Get feature importance from tree-based models
tree_models = ['Random Forest', 'Decision Tree']
feature_names = X_clean.columns

plt.figure(figsize=(15, 6))

for i, model_name in enumerate(tree_models, 1):
    if model_name in results:
        plt.subplot(1, 2, i)
        
        if hasattr(results[model_name]['model'], 'feature_importances_'):
            importances = results[model_name]['model'].feature_importances_
            indices = np.argsort(importances)[-10:]  # Top 10 features
            
            # Create horizontal bar chart
            plt.barh(range(len(indices)), importances[indices], color='purple', alpha=0.7)
            plt.yticks(range(len(indices)), [feature_names[i] for i in indices])
            plt.xlabel('Feature Importance')
            plt.title(f'Top 10 Features - {model_name}', fontweight='bold')
            
            # Add value labels
            for j, (importance, feature) in enumerate(zip(importances[indices], [feature_names[i] for i in indices])):
                plt.text(importance, j, f' {importance:.3f}', va='center', fontsize=9)

plt.tight_layout()
plt.show()

# For logistic regression, show coefficients
if 'Logistic Regression' in results:
    print("\nLogistic Regression Coefficients (Top 10):")
    lr_model = results['Logistic Regression']['model']
    coefficients = lr_model.coef_[0]
    feature_imp_df = pd.DataFrame({
        'Feature': feature_names,
        'Coefficient': coefficients,
        'Abs_Coefficient': np.abs(coefficients)
    }).sort_values('Abs_Coefficient', ascending=False)
    
    print(feature_imp_df.head(10).to_string(index=False))
```

Explanation: Feature importance plots reveal which variables most influence model predictions. Tree-based models show inherent importance scores, while logistic regression uses coefficient magnitudes. Understanding these relationships helps validate model logic and identify key predictive factors in the data.

```python
# VISUALIZATION 5: Cross-Validation Performance
print("\n" + "="*50)
print("VISUALIZATION 5: Cross-Validation Performance")
print("="*50)

# Perform cross-validation for all models
cv_results = {}
for name, model in models.items():
    cv_scores = cross_val_score(model, X_clean, y_clean, cv=5, scoring='accuracy')
    cv_results[name] = cv_scores

plt.figure(figsize=(12, 6))

# Box plot for cross-validation results
cv_df = pd.DataFrame(cv_results)
plt.subplot(1, 2, 1)
box = plt.boxplot(cv_df.values, labels=cv_df.columns, patch_artist=True)
plt.title('Cross-Validation Performance Distribution', fontweight='bold')
plt.xticks(rotation=45)
plt.ylabel('Accuracy Score')

# Color the boxes
colors = ['lightblue', 'lightgreen', 'lightcoral', 'lightyellow', 'lightgray']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# Mean CV scores bar chart
plt.subplot(1, 2, 2)
mean_cv_scores = [cv_scores.mean() for cv_scores in cv_results.values()]
std_cv_scores = [cv_scores.std() for cv_scores in cv_results.values()]

bars = plt.bar(model_names, mean_cv_scores, yerr=std_cv_scores, 
               capsize=5, alpha=0.7, color=colors)
plt.ylabel('Mean CV Accuracy')
plt.title('Mean Cross-Validation Scores with Error Bars', fontweight='bold')
plt.xticks(rotation=45)

# Add value labels
for bar, mean_score in zip(bars, mean_cv_scores):
    plt.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.01,
             f'{mean_score:.3f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

print("Cross-Validation Results Summary:")
for name in model_names:
    print(f"{name}: {cv_results[name].mean():.4f} (+/- {cv_results[name].std() * 2:.4f})")
```

Explanation: Cross-validation results demonstrate model stability and generalization capability. The box plots show performance distribution across different data splits, while the bar chart with error bars indicates average performance and variability. Lower variability suggests more reliable model performance.

```python
# VISUALIZATION 6: Performance Summary Dashboard
print("\n" + "="*50)
print("VISUALIZATION 6: Performance Summary Dashboard")
print("="*50)

# Create a comprehensive summary plot
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

# 1. Model ranking
accuracy_scores = [results[name]['accuracy'] for name in model_names]
sorted_indices = np.argsort(accuracy_scores)[::-1]
sorted_models = [model_names[i] for i in sorted_indices]
sorted_scores = [accuracy_scores[i] for i in sorted_indices]

ax1.barh(sorted_models, sorted_scores, color='skyblue')
ax1.set_xlabel('Accuracy')
ax1.set_title('Model Performance Ranking', fontweight='bold')
ax1.set_xlim(0, 1.0)

# 2. Metric comparison for best model
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
best_scores = [results[best_model_name]['accuracy'], 
               results[best_model_name]['precision'],
               results[best_model_name]['recall'], 
               results[best_model_name]['f1']]

ax2.bar(metrics, best_scores, color=['blue', 'green', 'orange', 'red'], alpha=0.7)
ax2.set_ylabel('Score')
ax2.set_title(f'Best Model Metrics: {best_model_name}', fontweight='bold')
ax2.set_ylim(0, 1.0)

# 3. Training vs Test performance (if we had training scores)
# For demonstration, we'll use cross-validation means as proxy
cv_means = [cv_results[name].mean() for name in model_names]
test_accuracies = [results[name]['accuracy'] for name in model_names]

x = np.arange(len(model_names))
width = 0.35
ax3.bar(x - width/2, cv_means, width, label='CV Mean', alpha=0.7)
ax3.bar(x + width/2, test_accuracies, width, label='Test Set', alpha=0.7)
ax3.set_xlabel('Models')
ax3.set_ylabel('Accuracy')
ax3.set_title('Cross-Validation vs Test Set Performance', fontweight='bold')
ax3.set_xticks(x)
ax3.set_xticklabels(model_names, rotation=45)
ax3.legend()

# 4. Performance gaps
performance_gaps = [test_accuracies[i] - cv_means[i] for i in range(len(model_names))]
colors = ['green' if gap >= 0 else 'red' for gap in performance_gaps]
ax4.bar(model_names, performance_gaps, color=colors, alpha=0.7)
ax4.axhline(y=0, color='black', linestyle='-', alpha=0.3)
ax4.set_ylabel('Performance Gap (Test - CV)')
ax4.set_title('Generalization Performance Gaps', fontweight='bold')
ax4.set_xticklabels(model_names, rotation=45)

plt.tight_layout()
plt.show()

# Final summary
print(f"\nFINAL SUMMARY:")
print(f"• Best Model: {best_model_name} (Accuracy: {results[best_model_name]['accuracy']:.3f})")
print(f"• Most Consistent: {min(cv_results, key=lambda x: cv_results[x].std())}")
print(f"• Recommended for Deployment: {best_model_name}")
print(f"• Key Strengths: {results[best_model_name]['f1']:.3f} F1-Score")
```

Explanation: This comprehensive dashboard provides a holistic view of model performance. It includes performance rankings, detailed metrics for the best model, cross-validation vs test set comparisons, and generalization gaps. This final visualization helps stakeholders make informed decisions about model selection and deployment.
